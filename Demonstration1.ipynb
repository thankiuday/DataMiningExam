```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
X, y = make_moons(n_samples=300, noise=0.1, random_state=0) # generate datasets, then 300 samples,

scaler = StandardScaler() #transform the data
X_scaled = scaler.fit_transform(X) 
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=5) #Dbscan algoruth, eps means radious of dots, min_sample means neighbours around cluster
clusters = dbscan.fit_predict(X_scaled) # fit the model
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='plasma', s=50)
plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```
